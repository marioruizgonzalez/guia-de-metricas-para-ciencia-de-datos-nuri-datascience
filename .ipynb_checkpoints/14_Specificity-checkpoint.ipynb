# Specificity (Especificidad): La Métrica que Mide tu Capacidad para Evitar Falsos Positivos

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.datasets import make_classification
import pandas as pd

# Configuración para reproducibilidad y estilo
np.random.seed(42)
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("=== ANÁLISIS DE ESPECIFICIDAD Y SENSIBILIDAD EN DETECCIÓN DE FRAUDE ===\n")
```

```python
# 1. CREAR DATASET SINTÉTICO (SIMULACIÓN DE DETECCIÓN DE FRAUDE)
X, y = make_classification(
    n_samples=5000,
    n_features=10,
    n_informative=8,
    n_redundant=2,
    n_clusters_per_class=1,
    weights=[0.995, 0.005],  # 99.5% legítimas, 0.5% fraude
    flip_y=0.01,
    random_state=42
)

# División de los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("=== DISTRIBUCIÓN DE CLASES ===")
print(f"Clase 0 (Legítima): {np.sum(y_train == 0)} ({np.mean(y_train == 0):.2%})")
print(f"Clase 1 (Fraude):   {np.sum(y_train == 1)} ({np.mean(y_train == 1):.2%})")
```

```python
# 2. DEFINIR MODELOS A COMPARAR
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000)
}

# Función para calcular la especificidad
def calculate_specificity(y_true, y_pred):
    """
    Calcula la especificidad manualmente usando la matriz de confusión.
    Specificity = TN / (TN + FP)
    """
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return tn / (tn + fp)
```

```python
# 3. ENTRENAR MODELOS Y VISUALIZAR MATRICES DE CONFUSIÓN
results = {}
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for idx, (name, model) in enumerate(models.items()):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    specificity = calculate_specificity(y_test, y_pred)
    
    results[name] = {
        'TN': tn, 'FP': fp, 'FN': fn, 'TP': tp,
        'Specificity': specificity,
        'Sensitivity': tp / (tp + fn),
        'Precision': tp / (tp + fp) if (tp + fp) > 0 else 0
    }
    
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Legítima', 'Fraude'],
                yticklabels=['Legítima', 'Fraude'],
                ax=axes[idx])
    axes[idx].set_title(f'{name}\nSpecificity: {specificity:.4f}')
    axes[idx].set_xlabel('Predicción')
    axes[idx].set_ylabel('Valor Real')

plt.tight_layout()
plt.show()
```

```python
# 4. MOSTRAR RESULTADOS DETALLADOS
print("=== COMPARACIÓN DETALLADA DE ESPECIFICIDAD ===\n")

for name, metrics in results.items():
    print(f"--- {name} ---")
    print(f"True Negatives (TN):  {metrics['TN']:,}")
    print(f"False Positives (FP): {metrics['FP']:,}")
    print(f"Specificity:          {metrics['Specificity']:.4f} ({metrics['Specificity']*100:.2f}%)")
    print(f"Sensitivity (Recall): {metrics['Sensitivity']:.4f}")
    print(f"Precision:            {metrics['Precision']:.4f}")
    
    total_negatives = metrics['TN'] + metrics['FP']
    print(f"\nInterpretación práctica:")
    print(f"• De {total_negatives:,} transacciones legítimas reales:")
    print(f"  - {metrics['TN']:,} fueron correctamente identificadas como legítimas")
    print(f"  - {metrics['FP']:,} fueron incorrectamente marcadas como fraude")
    print(f"• Esto implica {metrics['FP']:,} falsos positivos (clientes molestados innecesariamente).")
    print("-" * 50)
```

```python
# 5. COMPARACIÓN VISUAL DE ESPECIFICIDAD, SENSIBILIDAD Y PRECISIÓN
metrics_comparison = pd.DataFrame(results).T

fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(metrics_comparison.index))
width = 0.25

ax.bar(x - width, metrics_comparison['Specificity'], width, label='Specificity', alpha=0.8, color='skyblue')
ax.bar(x, metrics_comparison['Sensitivity'], width, label='Sensitivity (Recall)', alpha=0.8, color='lightcoral')
ax.bar(x + width, metrics_comparison['Precision'], width, label='Precision', alpha=0.8, color='lightgreen')

ax.set_xlabel('Modelos')
ax.set_ylabel('Puntuación')
ax.set_title('Comparación de Métricas: Specificity vs Sensitivity vs Precision')
ax.set_xticks(x)
ax.set_xticklabels(metrics_comparison.index)
ax.legend()
ax.grid(True, alpha=0.3)

# Añadir valores sobre las barras
for i, model in enumerate(metrics_comparison.index):
    ax.text(i - width, metrics_comparison.loc[model, 'Specificity'] + 0.01, 
            f'{metrics_comparison.loc[model, "Specificity"]:.3f}', ha='center', va='bottom', fontsize=9)
    ax.text(i, metrics_comparison.loc[model, 'Sensitivity'] + 0.01, 
            f'{metrics_comparison.loc[model, "Sensitivity"]:.3f}', ha='center', va='bottom', fontsize=9)
    ax.text(i + width, metrics_comparison.loc[model, 'Precision'] + 0.01, 
            f'{metrics_comparison.loc[model, "Precision"]:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
# 6. IMPACTO DEL THRESHOLD EN LA ESPECIFICIDAD
print("\n=== IMPACTO DEL THRESHOLD EN ESPECIFICIDAD ===")

rf_model = models['Random Forest']
y_proba = rf_model.predict_proba(X_test)[:, 1]

thresholds = [0.1, 0.3, 0.5, 0.7, 0.9]
threshold_results = []

for threshold in thresholds:
    y_pred_threshold = (y_proba >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_threshold).ravel()
    spec = tn / (tn + fp)
    sens = tp / (tp + fn)
    
    threshold_results.append({
        'Threshold': threshold,
        'Specificity': spec,
        'Sensitivity': sens,
        'False_Positives': fp,
        'True_Negatives': tn
    })
    
    print(f"Threshold {threshold:.1f}: Specificity={spec:.4f}, Sensitivity={sens:.4f}, FP={fp}")

threshold_df = pd.DataFrame(threshold_results)
```

```python
# 7. GRÁFICOS DE TRADE-OFF ENTRE MÉTRICAS Y FALSOS POSITIVOS
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# (a) Specificity vs Sensitivity
ax1.plot(threshold_df['Threshold'], threshold_df['Specificity'], marker='o', label='Specificity', linewidth=2, color='blue')
ax1.plot(threshold_df['Threshold'], threshold_df['Sensitivity'], marker='s', label='Sensitivity', linewidth=2, color='red')
ax1.set_xlabel('Threshold')
ax1.set_ylabel('Puntuación')
ax1.set_title('Trade-off: Specificity vs Sensitivity por Threshold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# (b) Falsos positivos por threshold
ax2.bar(threshold_df['Threshold'], threshold_df['False_Positives'], alpha=0.7, color='orange')
ax2.set_xlabel('Threshold')
ax2.set_ylabel('Número de Falsos Positivos')
ax2.set_title('Falsos Positivos por Threshold')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
