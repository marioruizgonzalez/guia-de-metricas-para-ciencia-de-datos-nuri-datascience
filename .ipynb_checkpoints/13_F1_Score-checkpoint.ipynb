# F1 Score: El Equilibrio Perfecto entre Precisión y Recall

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    f1_score, precision_score, recall_score, confusion_matrix,
    classification_report, precision_recall_curve
)
import pandas as pd

# Configuración general
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
np.random.seed(42)

print("=== ANÁLISIS COMPLETO DE F1 SCORE EN DETECCIÓN DE FRAUDE ===\n")
```

```python
# 1. CREAR DATASET DESBALANCEADO (SIMULACIÓN DE FRAUDE)
print("=== CREANDO DATASET DESBALANCEADO ===")

X, y = make_classification(
    n_samples=10000,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    n_classes=2,
    weights=[0.99, 0.01],  # 1% fraudes
    flip_y=0.01,
    random_state=42
)

# División train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Distribución de clases en entrenamiento:")
print(f"Clase 0 (Normal): {np.sum(y_train == 0)} ({np.mean(y_train == 0):.1%})")
print(f"Clase 1 (Fraude): {np.sum(y_train == 1)} ({np.mean(y_train == 1):.1%})")
```

```python
# 2. ENTRENAR DIFERENTES MODELOS
print("\n=== ENTRENANDO MODELOS ===")

models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000)
}

results = {}

for name, model in models.items():
    print(f"\nEntrenando {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calcular métricas principales
    f1 = f1_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    
    results[name] = {
        'F1': f1,
        'Precision': precision,
        'Recall': recall,
        'y_pred': y_pred,
        'model': model
    }
    
    print(f"F1 Score: {f1:.3f}")
    print(f"Precision: {precision:.3f}")
    print(f"Recall: {recall:.3f}")
```

```python
# 3. VISUALIZACIONES COMPARATIVAS
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Análisis Comparativo del F1 Score', fontsize=16, fontweight='bold')

# Gráfico 1: Comparación de métricas
ax1 = axes[0, 0]
metrics_df = pd.DataFrame(results).T
metrics_df[['F1', 'Precision', 'Recall']].plot(kind='bar', ax=ax1)
ax1.set_title('Comparación de Métricas por Modelo')
ax1.set_ylabel('Score')
ax1.legend()
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

# Gráfico 2: Matriz de confusión del mejor modelo
best_model_name = max(results.keys(), key=lambda k: results[k]['F1'])
best_y_pred = results[best_model_name]['y_pred']

ax2 = axes[0, 1]
cm = confusion_matrix(y_test, best_y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax2)
ax2.set_title(f'Matriz de Confusión - {best_model_name}')
ax2.set_ylabel('Actual')
ax2.set_xlabel('Predicción')

# Gráfico 3: Curvas Precision-Recall
ax3 = axes[1, 0]
for name, result in results.items():
    model = result['model']
    y_proba = model.predict_proba(X_test)[:, 1]
    precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_proba)
    ax3.plot(recall_curve, precision_curve, label=f'{name} (F1={result["F1"]:.3f})')

ax3.set_xlabel('Recall')
ax3.set_ylabel('Precision')
ax3.set_title('Curva Precision-Recall')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Gráfico 4: Análisis de umbrales
ax4 = axes[1, 1]
model = results[best_model_name]['model']
y_proba = model.predict_proba(X_test)[:, 1]

thresholds = np.linspace(0, 1, 100)
f1_scores, precisions, recalls = [], [], []

for threshold in thresholds:
    y_pred_thresh = (y_proba >= threshold).astype(int)
    if len(np.unique(y_pred_thresh)) > 1:  # evitar divisiones por cero
        f1_scores.append(f1_score(y_test, y_pred_thresh))
        precisions.append(precision_score(y_test, y_pred_thresh))
        recalls.append(recall_score(y_test, y_pred_thresh))
    else:
        f1_scores.append(0)
        precisions.append(0)
        recalls.append(0)

ax4.plot(thresholds, f1_scores, label='F1 Score', linewidth=2)
ax4.plot(thresholds, precisions, label='Precision', alpha=0.7)
ax4.plot(thresholds, recalls, label='Recall', alpha=0.7)
ax4.set_xlabel('Threshold')
ax4.set_ylabel('Score')
ax4.set_title('Métricas vs Threshold de Decisión')
ax4.legend()
ax4.grid(True, alpha=0.3)

# Threshold óptimo (máximo F1)
optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]
ax4.axvline(optimal_threshold, color='red', linestyle='--', label=f'Óptimo F1 (t={optimal_threshold:.2f})')
ax4.legend()

plt.tight_layout()
plt.show()

```

```python
# 4. REPORTE DETALLADO
print("\n" + "="*50)
print("REPORTE DETALLADO DE F1 SCORE")
print("="*50)

print(f"\nMejor modelo: {best_model_name}")
print(f"F1 Score: {results[best_model_name]['F1']:.4f}")
print(f"Threshold óptimo: {optimal_threshold:.3f}")
print(f"F1 máximo posible: {max(f1_scores):.4f}")

print("\nReporte de clasificación completo:")
print(classification_report(y_test, best_y_pred, target_names=['Normal', 'Fraude']))
```

```python
# 5. ANÁLISIS DE TRADE-OFFS ENTRE ESTRATEGIAS
print(f"\n--- ANÁLISIS DE TRADE-OFFS ---")

model = results[best_model_name]['model']
y_proba = model.predict_proba(X_test)[:, 1]

strategies = {
    'Alta Precisión': 0.8,
    'Balanced (F1)': optimal_threshold,
    'Alto Recall': 0.3
}

print(f"{'Estrategia':<15} {'Threshold':<10} {'F1':<6} {'Precision':<10} {'Recall':<8}")
print("-" * 50)

for strategy, threshold in strategies.items():
    y_pred_strategy = (y_proba >= threshold).astype(int)
    if len(np.unique(y_pred_strategy)) > 1:
        f1 = f1_score(y_test, y_pred_strategy)
        precision = precision_score(y_test, y_pred_strategy)
        recall = recall_score(y_test, y_pred_strategy)
        print(f"{strategy:<15} {threshold:<10.3f} {f1:<6.3f} {precision:<10.3f} {recall:<8.3f}")
```

```python
print(f"\n--- INTERPRETACIÓN PRÁCTICA ---")
print(f"• Con threshold alto (0.8): Muy pocos falsos positivos, pero perdemos muchos fraudes.")
print(f"• Con threshold óptimo F1 ({optimal_threshold:.2f}): Balance entre detectar fraudes y evitar falsas alarmas.")
print(f"• Con threshold bajo (0.3): Detectamos casi todos los fraudes, pero con muchas falsas alarmas.")
```
