# Balanced Accuracy: Evaluación Equilibrada para Datasets Desbalanceados

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    balanced_accuracy_score, accuracy_score, 
    classification_report, confusion_matrix,
    precision_recall_fscore_support
)

# Configuración general
np.random.seed(42)
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("=== ANÁLISIS DE BALANCED ACCURACY EN CLASIFICACIÓN DESBALANCEADA ===\n")
```

```python
# 1. CREAR DATASET DESBALANCEADO
X, y = make_classification(
    n_samples=10000,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    n_classes=2,
    weights=[0.95, 0.05],  # Altamente desbalanceado (95% vs 5%)
    random_state=42
)

# División train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("=== DISTRIBUCIÓN DEL DATASET ===")
print(f"Clase 0: {np.sum(y_train == 0)} muestras ({np.mean(y_train == 0)*100:.1f}%)")
print(f"Clase 1: {np.sum(y_train == 1)} muestras ({np.mean(y_train == 1)*100:.1f}%)")
```

```python
# 2. DEFINIR Y ENTRENAR MODELOS
models = {
    'Random Forest (sin balancear)': RandomForestClassifier(random_state=42),
    'Random Forest (balanceado)': RandomForestClassifier(class_weight='balanced', random_state=42),
    'Logistic Regression (sin balancear)': LogisticRegression(random_state=42, max_iter=1000),
    'Logistic Regression (balanceado)': LogisticRegression(class_weight='balanced', random_state=42)
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    results[name] = {
        'accuracy': accuracy_score(y_test, y_pred),
        'balanced_accuracy': balanced_accuracy_score(y_test, y_pred),
        'predictions': y_pred
    }
```

```python
# 3. MOSTRAR COMPARACIÓN DE MÉTRICAS
print("\n" + "="*70)
print("COMPARACIÓN DE MÉTRICAS")
print("="*70)
print(f"{'Modelo':<35} {'Accuracy':<12} {'Balanced Acc':<15}")
print("-"*70)

for name, metrics in results.items():
    print(f"{name:<35} {metrics['accuracy']:<12.4f} {metrics['balanced_accuracy']:<15.4f}")
```

```python
# 4. ANÁLISIS DETALLADO DEL MEJOR MODELO BALANCEADO
best_model_name = 'Random Forest (balanceado)'
best_predictions = results[best_model_name]['predictions']

print(f"\n\nANÁLISIS DETALLADO: {best_model_name}")
print("="*50)
print(classification_report(
    y_test, best_predictions, 
    target_names=['Clase 0 (Mayoritaria)', 'Clase 1 (Minoritaria)']
))
```

```python
# 5. VISUALIZACIONES COMPARATIVAS
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Análisis del Balanced Accuracy', fontsize=16, fontweight='bold')

# (1) Comparación Accuracy vs Balanced Accuracy
model_names = list(results.keys())
accuracies = [results[name]['accuracy'] for name in model_names]
balanced_accuracies = [results[name]['balanced_accuracy'] for name in model_names]

x = np.arange(len(model_names))
width = 0.35

ax1.bar(x - width/2, accuracies, width, label='Accuracy', alpha=0.8)
ax1.bar(x + width/2, balanced_accuracies, width, label='Balanced Accuracy', alpha=0.8)
ax1.set_xlabel('Modelos')
ax1.set_ylabel('Score')
ax1.set_title('Comparación: Accuracy vs Balanced Accuracy')
ax1.set_xticks(x)
ax1.set_xticklabels([name.replace(' ', '\n') for name in model_names], fontsize=8)
ax1.legend()
ax1.grid(True, alpha=0.3)

# (2) Matriz de confusión del mejor modelo
cm = confusion_matrix(y_test, best_predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax2)
ax2.set_title(f'Matriz de Confusión\n{best_model_name}')
ax2.set_xlabel('Predicción')
ax2.set_ylabel('Valor Real')

# (3) Distribución de clases en test set
class_counts = np.bincount(y_test)
ax3.pie(class_counts, labels=['Clase 0', 'Clase 1'], autopct='%1.1f%%', startangle=90)
ax3.set_title('Distribución de Clases en Test Set')

# (4) Métricas por clase
precision, recall, f1, _ = precision_recall_fscore_support(y_test, best_predictions)
classes = ['Clase 0', 'Clase 1']
metrics_data = {'Precision': precision, 'Recall': recall, 'F1-Score': f1}

x = np.arange(len(classes))
width = 0.25

for i, (metric, values) in enumerate(metrics_data.items()):
    ax4.bar(x + i*width, values, width, label=metric, alpha=0.8)

ax4.set_xlabel('Clases')
ax4.set_ylabel('Score')
ax4.set_title('Métricas por Clase (Mejor Modelo)')
ax4.set_xticks(x + width)
ax4.set_xticklabels(classes)
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# 6. CÁLCULO MANUAL DE BALANCED ACCURACY
print("\n" + "="*50)
print("CÁLCULO MANUAL DE BALANCED ACCURACY")
print("="*50)

tn, fp, fn, tp = confusion_matrix(y_test, best_predictions).ravel()

print(f"True Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
print(f"True Positives (TP): {tp}")

# Sensibilidad (clase positiva) y especificidad (clase negativa)
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print(f"\nSensitivity (Recall clase 1): {sensitivity:.4f}")
print(f"Specificity (Recall clase 0): {specificity:.4f}")

# Balanced Accuracy
print(f"Balanced Accuracy manual: {(sensitivity + specificity) / 2:.4f}")
print(f"Balanced Accuracy sklearn: {balanced_accuracy_score(y_test, best_predictions):.4f}")
```
