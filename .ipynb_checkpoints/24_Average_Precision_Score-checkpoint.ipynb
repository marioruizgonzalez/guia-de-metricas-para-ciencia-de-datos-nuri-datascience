# Average Precision Score: Guía Completa para Evaluación de Modelos de ciencia de datos

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    average_precision_score,
    precision_recall_curve,
    classification_report,
    confusion_matrix
)

# Configuración estética
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
np.random.seed(42)

print("=== ANÁLISIS DE CURVAS PRECISION-RECALL Y AVERAGE PRECISION SCORE ===\n")
```

```python
X, y = make_classification(
    n_samples=5000, 
    n_features=20, 
    n_informative=15,
    n_redundant=5,
    n_clusters_per_class=1,
    weights=[0.95, 0.05],
    random_state=42
)

print(f"Distribución de clases: {np.bincount(y)}")
print(f"Porcentaje de clase positiva (Fraude): {y.mean():.2%}")

# División train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Calcular métricas PR
    precision, recall, thresholds = precision_recall_curve(y_test, y_proba)
    ap_score = average_precision_score(y_test, y_proba)

    results[name] = {
        'ap_score': ap_score,
        'precision': precision,
        'recall': recall,
        'y_proba': y_proba
    }

    print(f"{name:20}: Average Precision Score = {ap_score:.4f}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
baseline_ap = y_test.mean()

# Subplot 1 — Curvas Precision-Recall
ax1 = axes[0, 0]
for name, data in results.items():
    ax1.plot(data['recall'], data['precision'],
             label=f"{name} (AP={data['ap_score']:.3f})", linewidth=2)
ax1.axhline(y=baseline_ap, color='red', linestyle='--', 
            label=f'Baseline (AP={baseline_ap:.3f})')
ax1.set_xlabel('Recall')
ax1.set_ylabel('Precision')
ax1.set_title('Curvas Precision-Recall')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Subplot 2 — Distribución de probabilidades (Random Forest)
ax2 = axes[0, 1]
rf_proba = results['Random Forest']['y_proba']
ax2.hist(rf_proba[y_test == 0], bins=30, alpha=0.7, color='lightblue', 
         label='Clase Negativa', density=True)
ax2.hist(rf_proba[y_test == 1], bins=30, alpha=0.7, color='orange', 
         label='Clase Positiva', density=True)
ax2.set_xlabel('Probabilidad Predicha')
ax2.set_ylabel('Densidad')
ax2.set_title('Distribución de Probabilidades - Random Forest')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Subplot 3 — Comparación de AP Scores
ax3 = axes[1, 0]
ap_scores = [data['ap_score'] for data in results.values()]
model_names = list(results.keys())
bars = ax3.bar(model_names, ap_scores, color=['skyblue', 'lightcoral'])
ax3.axhline(y=baseline_ap, color='red', linestyle='--', 
            label=f'Baseline ({baseline_ap:.3f})')
for bar, score in zip(bars, ap_scores):
    ax3.text(bar.get_x() + bar.get_width()/2., score + 0.01,
             f'{score:.3f}', ha='center', va='bottom', fontweight='bold')
ax3.set_ylabel('Average Precision Score')
ax3.set_title('Comparación de Average Precision Scores')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Subplot 4 — Análisis de umbrales
ax4 = axes[1, 1]
rf_precision = results['Random Forest']['precision']
rf_recall = results['Random Forest']['recall']
f1_scores = 2 * (rf_precision[:-1] * rf_recall[:-1]) / (rf_precision[:-1] + rf_recall[:-1])
precision, recall, thresholds = precision_recall_curve(y_test, rf_proba)
ax4.plot(thresholds, precision[:-1], label='Precision', linewidth=2)
ax4.plot(thresholds, recall[:-1], label='Recall', linewidth=2)
ax4.plot(thresholds, f1_scores, label='F1-Score', linewidth=2)
ax4.set_xlabel('Umbral')
ax4.set_ylabel('Score')
ax4.set_title('Métricas vs Umbral de Decisión')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
best_model_name = max(results.keys(), key=lambda k: results[k]['ap_score'])
best_ap = results[best_model_name]['ap_score']

print("="*60)
print(f"ANÁLISIS DETALLADO - {best_model_name.upper()}")
print("="*60)
print(f"Average Precision Score: {best_ap:.4f}")
print(f"Mejora sobre baseline: {(best_ap - baseline_ap)/baseline_ap:.1%}")

# Calcular umbral óptimo (máximo F1)
rf_precision, rf_recall, rf_thresholds = precision_recall_curve(
    y_test, results[best_model_name]['y_proba']
)
f1_scores = 2 * (rf_precision[:-1] * rf_recall[:-1]) / (rf_precision[:-1] + rf_recall[:-1])
optimal_idx = np.argmax(f1_scores)
optimal_threshold = rf_thresholds[optimal_idx]

print(f"Umbral óptimo (F1): {optimal_threshold:.3f}")
print(f"Precision en umbral óptimo: {rf_precision[optimal_idx]:.3f}")
print(f"Recall en umbral óptimo: {rf_recall[optimal_idx]:.3f}")
print(f"F1-Score máximo: {f1_scores[optimal_idx]:.3f}")
```

```python
print("\n" + "="*60)
print("INTERPRETACIÓN PRÁCTICA")
print("="*60)
print(f"Con AP = {best_ap:.3f}, este modelo:")
print(f"• Mantiene alta precisión detectando aproximadamente {best_ap*100:.0f}% de los casos positivos.")
print(f"• Es {(best_ap - baseline_ap)/baseline_ap:.1f}x mejor que un clasificador aleatorio.")
print(f"• En un ranking de casos sospechosos, los verdaderos positivos")
print(f"  tienden a aparecer en las primeras posiciones (mayor score).")

if best_ap > 0.8:
    print("• EXCELENTE: Modelo muy confiable para producción.")
elif best_ap > 0.6:
    print("• BUENO: Modelo útil, considerar mejoras.")
else:
    print("• MEJORABLE: Revisar features, algoritmo o calidad de datos.")
```
