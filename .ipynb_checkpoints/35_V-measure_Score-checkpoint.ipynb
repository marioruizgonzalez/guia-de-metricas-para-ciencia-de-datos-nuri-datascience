# V-measure Score: La Métrica Equilibrada para Clustering

```python
import os, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import (
    v_measure_score,
    homogeneity_score,
    completeness_score,
    adjusted_rand_score,
    silhouette_score,
    confusion_matrix
)

# Configuración general
os.environ["OMP_NUM_THREADS"] = "2"
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak")

np.random.seed(42)
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("=== MÉTRICA: V-MEASURE SCORE ===\n")
```

```python
# Parámetros del dataset
n_samples = 500
n_centers = 4
n_features = 2

# Generar datos
X, y_true = make_blobs(
    n_samples=n_samples,
    centers=n_centers,
    n_features=n_features,
    random_state=42,
    cluster_std=1.5
)

print(f"Dataset creado: {n_samples} muestras, {n_centers} clusters verdaderos.")
print(f"Distribución de clases verdaderas: {np.bincount(y_true)}")
```

```python
k_range = range(2, 8)
results = []

print("\n=== EVALUACIÓN DE K-MEANS PARA DIFERENTES VALORES DE K ===")

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    y_pred = kmeans.fit_predict(X)
    
    # Calcular métricas principales
    v_measure = v_measure_score(y_true, y_pred)
    homogeneity = homogeneity_score(y_true, y_pred)
    completeness = completeness_score(y_true, y_pred)
    silhouette = silhouette_score(X, y_pred)
    ari = adjusted_rand_score(y_true, y_pred)
    
    results.append({
        'k': k,
        'v_measure': v_measure,
        'homogeneity': homogeneity,
        'completeness': completeness,
        'silhouette': silhouette,
        'ari': ari
    })
    
    print(f"K={k}: V-measure={v_measure:.3f}, Homogeneity={homogeneity:.3f}, Completeness={completeness:.3f}")

# Convertir resultados a DataFrame
df_results = pd.DataFrame(results)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Evaluación de K-Means con V-Measure Score', fontsize=16, fontweight='bold')

# --- (1) Comparación entre V-measure, Homogeneity y Completeness ---
ax1 = axes[0, 0]
ax1.plot(df_results['k'], df_results['v_measure'], 'o-', label='V-measure', linewidth=2, markersize=8)
ax1.plot(df_results['k'], df_results['homogeneity'], 's-', label='Homogeneity', linewidth=2)
ax1.plot(df_results['k'], df_results['completeness'], '^-', label='Completeness', linewidth=2)
ax1.axvline(x=4, color='red', linestyle='--', alpha=0.7, label='K verdadero')
ax1.set_xlabel('Número de Clusters (K)')
ax1.set_ylabel('Score')
ax1.set_title('V-measure y sus Componentes')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(0, 1)

# --- (2) Comparación con otras métricas ---
ax2 = axes[0, 1]
ax2.plot(df_results['k'], df_results['v_measure'], 'o-', label='V-measure', linewidth=2)
ax2.plot(df_results['k'], df_results['ari'], 's-', label='Adjusted Rand Index', linewidth=2)
ax2.plot(df_results['k'], df_results['silhouette'], '^-', label='Silhouette Score', linewidth=2)
ax2.axvline(x=4, color='red', linestyle='--', alpha=0.7, label='K verdadero')
ax2.set_xlabel('Número de Clusters (K)')
ax2.set_ylabel('Score')
ax2.set_title('Comparación con Otras Métricas')
ax2.legend()
ax2.grid(True, alpha=0.3)

# --- (3) Clusters verdaderos ---
ax3 = axes[1, 0]
scatter1 = ax3.scatter(X[:, 0], X[:, 1], c=y_true, cmap='tab10', alpha=0.7, s=50)
ax3.set_title('Clusters Verdaderos')
ax3.set_xlabel('Feature 1')
ax3.set_ylabel('Feature 2')

# --- (4) Clustering predicho (K=4) ---
ax4 = axes[1, 1]
kmeans_optimal = KMeans(n_clusters=4, random_state=42)
y_pred_optimal = kmeans_optimal.fit_predict(X)
scatter2 = ax4.scatter(X[:, 0], X[:, 1], c=y_pred_optimal, cmap='tab10', alpha=0.7, s=50)
centers = kmeans_optimal.cluster_centers_
ax4.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, linewidths=3, label='Centroides')
ax4.set_title('Clusters Predichos (K=4)')
ax4.set_xlabel('Feature 1')
ax4.set_ylabel('Feature 2')
ax4.legend()

plt.tight_layout()
plt.show()
```

```python
print("\n" + "="*60)
print("ANÁLISIS DETALLADO DEL CLUSTERING ÓPTIMO (K=4)")
print("="*60)

v_opt = v_measure_score(y_true, y_pred_optimal)
h_opt = homogeneity_score(y_true, y_pred_optimal)
c_opt = completeness_score(y_true, y_pred_optimal)

print(f"V-measure Score: {v_opt:.4f}")
print(f"Homogeneity: {h_opt:.4f}")
print(f"Completeness: {c_opt:.4f}")
print(f"Verificación matemática: 2*({h_opt:.4f}*{c_opt:.4f})/({h_opt:.4f}+{c_opt:.4f}) = {2*h_opt*c_opt/(h_opt+c_opt):.4f}")
```

```python
conf_matrix = confusion_matrix(y_true, y_pred_optimal)

plt.figure(figsize=(8, 6))
sns.heatmap(
    conf_matrix,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=[f'Pred {i}' for i in range(4)],
    yticklabels=[f'True {i}' for i in range(4)]
)
plt.title('Matriz de Confusión - Clustering (K=4)', fontweight='bold')
plt.xlabel('Clusters Predichos')
plt.ylabel('Clusters Verdaderos')
plt.show()
```

```python
print("\n" + "="*60)
print("COMPARACIÓN CON CLUSTERING SUBÓPTIMO (K=2)")
print("="*60)

kmeans_bad = KMeans(n_clusters=2, random_state=42)
y_pred_bad = kmeans_bad.fit_predict(X)

v_bad = v_measure_score(y_true, y_pred_bad)
h_bad = homogeneity_score(y_true, y_pred_bad)
c_bad = completeness_score(y_true, y_pred_bad)

print(f"V-measure Score: {v_bad:.4f}")
print(f"Homogeneity: {h_bad:.4f}")
print(f"Completeness: {c_bad:.4f}")
print(f"Diferencia con el óptimo: {v_opt - v_bad:.4f}")
```
