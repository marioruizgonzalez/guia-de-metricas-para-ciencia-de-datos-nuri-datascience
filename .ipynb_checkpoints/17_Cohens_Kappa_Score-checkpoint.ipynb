# Cohen's Kappa Score: Evaluando Acuerdo Más Allá del Azar

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import cohen_kappa_score, accuracy_score, classification_report, confusion_matrix
from sklearn.dummy import DummyClassifier
import pandas as pd

# Configuración visual
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("=== COHEN'S KAPPA SCORE — ANÁLISIS DE CASO DESBALANCEADO ===\n")
```

```python
# 1. CREAR DATASET (SIMULACIÓN DE DETECCIÓN DE FRAUDE)
print("=" * 60)
print("COHEN'S KAPPA SCORE - EJEMPLO PRÁCTICO")
print("Escenario: Detección de Fraude Bancario")
print("=" * 60)

X, y = make_classification(
    n_samples=10000,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    n_clusters_per_class=1,
    weights=[0.99, 0.01],  # 99% clase 0, 1% clase 1 (muy desbalanceado)
    flip_y=0.01,
    random_state=42
)

# División train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("=== DISTRIBUCIÓN DE CLASES (TRAIN) ===")
print(f"Clase 0 (No Fraude): {np.sum(y_train == 0)} ({np.mean(y_train == 0):.1%})")
print(f"Clase 1 (Fraude):    {np.sum(y_train == 1)} ({np.mean(y_train == 1):.1%})\n")
```

```python
# 2. ENTRENAR DIFERENTES MODELOS
modelos = {
    'Dummy (Siempre Mayoría)': DummyClassifier(strategy='most_frequent'),
    'Dummy (Estratificado)': DummyClassifier(strategy='stratified'),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
}

resultados = []

for nombre, modelo in modelos.items():
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    kappa = cohen_kappa_score(y_test, y_pred)
    
    resultados.append({
        'Modelo': nombre,
        'Accuracy': accuracy,
        "Cohen's Kappa": kappa,
        'Predicciones': y_pred
    })
    
    print(f"{nombre}:")
    print(f"  Accuracy: {accuracy:.4f}")
    print(f"  Cohen's Kappa: {kappa:.4f}\n")

# Convertir resultados a DataFrame
df_resultados = pd.DataFrame(resultados)
```

```python
# 3. VISUALIZACIONES COMPARATIVAS
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("Cohen's Kappa vs Accuracy — Comparación de Métricas\nDataset Desbalanceado (1% Fraude)", 
             fontsize=16, fontweight='bold')

# (1) Comparación de métricas
ax1 = axes[0, 0]
x_pos = np.arange(len(df_resultados))
width = 0.35

ax1.bar(x_pos - width/2, df_resultados['Accuracy'], width, label='Accuracy', alpha=0.8, color='skyblue')
ax1.bar(x_pos + width/2, df_resultados["Cohen's Kappa"], width, label="Cohen's Kappa", alpha=0.8, color='lightcoral')

ax1.set_xlabel('Modelos')
ax1.set_ylabel('Score')
ax1.set_title("Accuracy vs Cohen's Kappa por Modelo")
ax1.set_xticks(x_pos)
ax1.set_xticklabels(df_resultados['Modelo'], rotation=45, ha='right')
ax1.legend()
ax1.grid(True, alpha=0.3)

# (2) Scatter Accuracy vs Kappa
ax2 = axes[0, 1]
scatter = ax2.scatter(df_resultados['Accuracy'], df_resultados["Cohen's Kappa"], 
                      s=100, alpha=0.7, c=range(len(df_resultados)), cmap='viridis')

for i, modelo in enumerate(df_resultados['Modelo']):
    ax2.annotate(modelo.replace(' ', '\n'),
                 (df_resultados['Accuracy'].iloc[i], df_resultados["Cohen's Kappa"].iloc[i]),
                 xytext=(5, 5), textcoords='offset points', fontsize=8)

ax2.set_xlabel('Accuracy')
ax2.set_ylabel("Cohen's Kappa")
ax2.set_title('Relación entre Accuracy y Cohen\'s Kappa')
ax2.grid(True, alpha=0.3)
ax2.plot([0, 1], [0, 1], 'r--', alpha=0.5, label='Accuracy = Kappa')
ax2.legend()

# (3) Matriz de confusión del mejor modelo
best_model_idx = df_resultados["Cohen's Kappa"].idxmax()
y_pred_best = resultados[best_model_idx]['Predicciones']

ax3 = axes[1, 0]
cm = confusion_matrix(y_test, y_pred_best)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3,
            xticklabels=['No Fraude', 'Fraude'],
            yticklabels=['No Fraude', 'Fraude'])
ax3.set_title(f"Matriz de Confusión - {resultados[best_model_idx]['Modelo']}")
ax3.set_xlabel('Predicción')
ax3.set_ylabel('Valor Real')

# (4) Escala interpretativa de Kappa
ax4 = axes[1, 1]
kappa_ranges = ['≤ 0\n(Pobre)', '0.01–0.20\n(Leve)', '0.21–0.40\n(Aceptable)',
                '0.41–0.60\n(Moderado)', '0.61–0.80\n(Sustancial)', '0.81–1.00\n(Casi Perfecto)']
kappa_values = [0, 0.1, 0.3, 0.5, 0.7, 0.9]
colors = ['red', 'orange', 'yellow', 'lightgreen', 'green', 'darkgreen']

ax4.bar(range(len(kappa_ranges)), kappa_values, color=colors, alpha=0.7)
ax4.set_xlabel("Interpretación de Cohen's Kappa")
ax4.set_ylabel("Valor Kappa")
ax4.set_title("Escala de Interpretación de Cohen's Kappa")
ax4.set_xticks(range(len(kappa_ranges)))
ax4.set_xticklabels(kappa_ranges, rotation=45, ha='right')

# Marcar los valores de Kappa de los modelos
for i, kappa_val in enumerate(df_resultados["Cohen's Kappa"]):
    ax4.axhline(y=kappa_val, color='red', linestyle='--', alpha=0.7)
    ax4.text(len(kappa_ranges)-1, kappa_val, f'{df_resultados["Modelo"].iloc[i][:10]}...',
             fontsize=8, ha='right', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# 4. ANÁLISIS DETALLADO DEL MEJOR MODELO
print("=" * 60)
print("ANÁLISIS DETALLADO DEL MEJOR MODELO")
print("=" * 60)

best_model_name = resultados[best_model_idx]['Modelo']
best_y_pred = resultados[best_model_idx]['Predicciones']

print(f"Mejor modelo según Cohen's Kappa: {best_model_name}")
print(f"Cohen's Kappa: {df_resultados['Cohen\'s Kappa'].iloc[best_model_idx]:.4f}")
print(f"Accuracy: {df_resultados['Accuracy'].iloc[best_model_idx]:.4f}\n")

print("Reporte de Clasificación:")
print(classification_report(y_test, best_y_pred, target_names=['No Fraude', 'Fraude']))
```

```python
# 5. CÁLCULO MANUAL DE KAPPA
print("=" * 60)
print("CÁLCULO MANUAL DE COHEN'S KAPPA")
print("=" * 60)

tn, fp, fn, tp = confusion_matrix(y_test, best_y_pred).ravel()

# P(observado)
p_observed = (tp + tn) / (tp + tn + fp + fn)

# P(esperado)
p_real_pos = (tp + fn) / (tp + tn + fp + fn)
p_pred_pos = (tp + fp) / (tp + tn + fp + fn)
p_real_neg = (tn + fp) / (tp + tn + fp + fn)
p_pred_neg = (tn + fn) / (tp + tn + fp + fn)
p_expected = (p_real_pos * p_pred_pos) + (p_real_neg * p_pred_neg)

# Cálculo manual
kappa_manual = (p_observed - p_expected) / (1 - p_expected)

print(f"TP={tp}, TN={tn}, FP={fp}, FN={fn}")
print(f"P(observado): {p_observed:.4f}")
print(f"P(esperado): {p_expected:.4f}")
print(f"Kappa manual: {kappa_manual:.4f}")
print(f"Kappa sklearn: {cohen_kappa_score(y_test, best_y_pred):.4f}")
```

```python
# 6. INTERPRETACIÓN PRÁCTICA
print("=" * 60)
print("INTERPRETACIÓN PRÁCTICA")
print("=" * 60)

kappa_best = df_resultados["Cohen's Kappa"].iloc[best_model_idx]

if kappa_best <= 0:
    interpretacion = "Pobre — sin acuerdo más allá del azar"
elif kappa_best <= 0.20:
    interpretacion = "Leve — acuerdo mínimo"
elif kappa_best <= 0.40:
    interpretacion = "Aceptable — bajo acuerdo"
elif kappa_best <= 0.60:
    interpretacion = "Moderado — acuerdo intermedio"
elif kappa_best <= 0.80:
    interpretacion = "Sustancial — buen acuerdo"
else:
    interpretacion = "Casi perfecto — excelente acuerdo"

print(f"Cohen's Kappa = {kappa_best:.4f} → {interpretacion}\n")

print("¿Por qué Kappa es mejor que Accuracy en este caso?")
print(f"• El dataset tiene {np.mean(y_test == 1):.1%} de fraudes reales (muy desbalanceado).")
acc_dummy = df_resultados[df_resultados['Modelo'] == 'Dummy (Siempre Mayoría)']['Accuracy'].iloc[0]
kap_dummy = df_resultados[df_resultados['Modelo'] == 'Dummy (Siempre Mayoría)']["Cohen's Kappa"].iloc[0]
print(f"• Un modelo que siempre predice 'No Fraude' tiene Accuracy = {acc_dummy:.1%}, pero Kappa = {kap_dummy:.4f}")
print("• Esto demuestra que la alta Accuracy no implica capacidad predictiva real.")
print("• El modelo con Kappa más alto muestra verdadero poder de discriminación incluso con desbalance extremo.")
```
