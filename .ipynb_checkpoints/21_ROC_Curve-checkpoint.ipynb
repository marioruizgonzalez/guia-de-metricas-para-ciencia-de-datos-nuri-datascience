# ROC Curve (Receiver Operating Characteristic)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc, classification_report, confusion_matrix
from sklearn.calibration import CalibratedClassifierCV

# Estilo visual uniforme
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
np.random.seed(42)

print("=== ANÁLISIS DE CURVAS ROC Y AUC ===\n")
```

```python
X, y = make_classification(
    n_samples=10_000,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    n_clusters_per_class=1,
    weights=[0.99, 0.01],  # 1% positivos (fraude)
    flip_y=0.01,
    random_state=42
)

print(f"Distribución de clases:")
print(f"  Clase 0 (No Fraude): {np.sum(y == 0)} ({np.mean(y == 0):.2%})")
print(f"  Clase 1 (Fraude):    {np.sum(y == 1)} ({np.mean(y == 1):.2%})")

# División entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
modelos = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM (Calibrated)': SVC(probability=True, random_state=42)
}

# Entrenamiento
for nombre, modelo in modelos.items():
    modelo.fit(X_train, y_train)
print("\nEntrenamiento completado en los 3 modelos.")
```

```python
resultados = {}
colores = ['#3498DB', '#E74C3C', '#2ECC71']

plt.figure(figsize=(12, 8))
for i, (nombre, modelo) in enumerate(modelos.items()):
    # Probabilidades de clase positiva
    y_proba = modelo.predict_proba(X_test)[:, 1]
    
    # Calcular curva ROC y AUC
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    
    resultados[nombre] = {'fpr': fpr, 'tpr': tpr, 'thresholds': thresholds, 
                          'auc': roc_auc, 'y_proba': y_proba}
    
    plt.plot(fpr, tpr, lw=2, color=colores[i], 
             label=f"{nombre} (AUC = {roc_auc:.3f})")

# Clasificador aleatorio
plt.plot([0, 1], [0, 1], 'k--', lw=2, alpha=0.5, label='Clasificador Aleatorio (AUC = 0.500)')

# Configuración general
plt.title('Comparación de Curvas ROC - Detección de Fraude', fontsize=14, fontweight='bold')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)

# Punto (0,1)
plt.plot(0, 1, 'ko', markersize=7)
plt.annotate('Clasificador Perfecto\n(0,1)', xy=(0, 1), xytext=(0.3, 0.9),
             arrowprops=dict(arrowstyle='->', color='black'), fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Selección del mejor modelo por AUC
mejor_modelo = max(resultados.keys(), key=lambda x: resultados[x]['auc'])
print(f"\n=== MEJOR MODELO: {mejor_modelo} ===")
print(f"AUC Score: {resultados[mejor_modelo]['auc']:.4f}")

fpr_mejor = resultados[mejor_modelo]['fpr']
tpr_mejor = resultados[mejor_modelo]['tpr']
thresholds_mejor = resultados[mejor_modelo]['thresholds']

# Umbral óptimo (distancia mínima a (0,1))
distancias = np.sqrt((fpr_mejor - 0)**2 + (tpr_mejor - 1)**2)
idx_opt = np.argmin(distancias)
umbral_optimo = thresholds_mejor[idx_opt]

print(f"\nUmbral óptimo: {umbral_optimo:.4f}")
print(f"TPR (Sensibilidad): {tpr_mejor[idx_opt]:.4f}")
print(f"FPR: {fpr_mejor[idx_opt]:.4f}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle(f"Análisis de Umbral — {mejor_modelo}", fontsize=14, fontweight='bold')

# (1) Sensibilidad del umbral
umbrales = thresholds_mejor[1:-1]
ax1.plot(umbrales, tpr_mejor[1:-1], 'b-', label='TPR (Sensibilidad)', linewidth=2)
ax1.plot(umbrales, fpr_mejor[1:-1], 'r-', label='FPR (1 - Especificidad)', linewidth=2)
ax1.axvline(umbral_optimo, color='green', linestyle='--', label=f'Umbral Óptimo = {umbral_optimo:.3f}')
ax1.set_xlabel('Umbral de Decisión')
ax1.set_ylabel('Tasa')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_title('TPR y FPR vs Umbral de Decisión')

# (2) Distribución de scores
y_proba = resultados[mejor_modelo]['y_proba']
ax2.hist(y_proba[y_test == 0], bins=50, color='blue', alpha=0.6, density=True, label='No Fraude')
ax2.hist(y_proba[y_test == 1], bins=50, color='red', alpha=0.6, density=True, label='Fraude')
ax2.axvline(umbral_optimo, color='green', linestyle='--', label='Umbral Óptimo')
ax2.set_xlabel('Score de Probabilidad')
ax2.set_ylabel('Densidad')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_title('Distribución de Scores por Clase')

plt.tight_layout()
plt.show()
```

```python
# Aplicar umbral óptimo
y_pred_optimo = (y_proba >= umbral_optimo).astype(int)
cm = confusion_matrix(y_test, y_pred_optimo)

print(f"\n=== MÉTRICAS CON UMBRAL ÓPTIMO ({umbral_optimo:.4f}) ===")
print("Matriz de Confusión:")
print(cm)

print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_optimo, target_names=['No Fraude', 'Fraude']))

# Cálculo manual
tn, fp, fn, tp = cm.ravel()
tpr = tp / (tp + fn)
fpr = fp / (fp + tn)
precision = tp / (tp + fp)
specificity = tn / (tn + fp)

print(f"\nTPR (Sensibilidad): {tpr:.4f}")
print(f"FPR: {fpr:.4f}")
print(f"Especificidad: {specificity:.4f}")
print(f"Precisión: {precision:.4f}")
```

```python
print("\n=== COMPARACIÓN FINAL DE AUC ===")
for nombre, datos in resultados.items():
    print(f"{nombre:20}: AUC = {datos['auc']:.4f}")
```
