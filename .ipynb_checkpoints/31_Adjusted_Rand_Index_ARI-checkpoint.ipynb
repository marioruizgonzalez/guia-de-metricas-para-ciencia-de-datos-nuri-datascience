# Adjusted Rand Index (ARI): Evaluando la Calidad de Clustering


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics.cluster import contingency_matrix
import os, warnings

os.environ["OMP_NUM_THREADS"] = "2"
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak")

# Configuración general
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
np.random.seed(42)

print("=== ANÁLISIS DE CLUSTERING USANDO ADJUSTED RAND INDEX ===\n")
```

```python
# Generar datos con 4 clusters reales
X, y_true = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=0.8,
    random_state=42,
    center_box=(-10.0, 10.0)
)

# Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"Datos generados: {X.shape[0]} muestras, {X.shape[1]} características.")
```

```python
algorithms = {
    'K-Means (k=4)': KMeans(n_clusters=4, random_state=42, n_init=10),
    'K-Means (k=3)': KMeans(n_clusters=3, random_state=42, n_init=10),
    'K-Means (k=5)': KMeans(n_clusters=5, random_state=42, n_init=10),
    'Hierarchical (n=4)': AgglomerativeClustering(n_clusters=4),
    'Hierarchical (n=3)': AgglomerativeClustering(n_clusters=3)
}

results = {}
predictions = {}

print("Evaluando modelos de clustering con ARI:\n")

for name, algorithm in algorithms.items():
    y_pred = algorithm.fit_predict(X_scaled)
    predictions[name] = y_pred
    ari_score = adjusted_rand_score(y_true, y_pred)
    results[name] = ari_score
    print(f"{name:22s} -> ARI = {ari_score:.3f}")
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

# (1) Clusters reales
axes[0].scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', alpha=0.7)
axes[0].set_title('Ground Truth\n(Clusters Reales)', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# (2–6) Predicciones de cada algoritmo
for i, (name, y_pred) in enumerate(predictions.items(), 1):
    ari = results[name]
    axes[i].scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', alpha=0.7)
    axes[i].set_title(f'{name}\nARI = {ari:.3f}', fontsize=12, fontweight='bold')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

algorithms_names = list(results.keys())
ari_scores = list(results.values())

colors = sns.color_palette("husl", len(ari_scores))
bars = plt.bar(algorithms_names, ari_scores, color=colors, alpha=0.8, edgecolor='black')

# Etiquetas en las barras
for bar, score in zip(bars, ari_scores):
    plt.text(
        bar.get_x() + bar.get_width()/2., 
        bar.get_height() + 0.02, 
        f'{score:.3f}', 
        ha='center', va='bottom', fontweight='bold'
    )

plt.title('Comparación de Algoritmos de Clustering usando ARI', fontsize=14, fontweight='bold', pad=20)
plt.ylabel('Adjusted Rand Index (ARI)', fontsize=12)
plt.xlabel('Algoritmo de Clustering', fontsize=12)
plt.ylim(0, 1.1)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.7, label='Concordancia Aleatoria')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
best_algorithm = max(results, key=results.get)
best_score = results[best_algorithm]

print("\n" + "="*55)
print("ANÁLISIS DEL MEJOR MODELO")
print("="*55)
print(f"Algoritmo: {best_algorithm}")
print(f"ARI Score: {best_score:.3f}\n")

print("Interpretación:")
if best_score > 0.8:
    print("• Excelente concordancia con la verdad fundamental")
elif best_score > 0.6:
    print("• Buena concordancia con la verdad fundamental")
elif best_score > 0.4:
    print("• Concordancia moderada con la verdad fundamental")
elif best_score > 0.0:
    print("• Concordancia baja pero mejor que el azar")
else:
    print("• Concordancia peor que una asignación aleatoria")
```

```python
print("\n" + "="*55)
print("EJEMPLO DE CÁLCULO MANUAL DEL ARI")
print("="*55)

# Tomar muestra aleatoria
sample_size = 10
indices = np.random.choice(len(X), sample_size, replace=False)
y_true_sample = y_true[indices]
y_pred_sample = predictions[best_algorithm][indices]

print(f"Muestra de {sample_size} puntos:")
print("Ground Truth:", y_true_sample)
print("Predicción:  ", y_pred_sample)

# Calcular matriz de contingencia
contingency = contingency_matrix(y_true_sample, y_pred_sample)
print("\nMatriz de Contingencia:")
print(contingency)

# ARI manual (para referencia)
ari_manual = adjusted_rand_score(y_true_sample, y_pred_sample)
print(f"\nARI calculado para la muestra: {ari_manual:.3f}")
```
