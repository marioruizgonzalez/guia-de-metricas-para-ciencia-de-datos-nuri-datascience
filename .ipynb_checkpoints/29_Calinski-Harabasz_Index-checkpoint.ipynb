# Calinski-Harabasz Index: Evaluando la Calidad de Clusters

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import calinski_harabasz_score
import warnings, os
warnings.filterwarnings("ignore", category=UserWarning)
os.environ["OMP_NUM_THREADS"] = "4"

# Configuración general
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
np.random.seed(42)

print("=== ANÁLISIS DE CLUSTERING CON CALINSKI–HARABASZ INDEX ===\n")
```

```python
def create_customer_data(n_samples=1000):
    """
    Simula datos de clientes para segmentación.
    Devuelve: X (datos), y_true (etiquetas verdaderas), nombres de features.
    """
    X, y_true = make_blobs(
        n_samples=n_samples,
        centers=4,
        n_features=2,
        cluster_std=1.5,
        random_state=42
    )
    feature_names = ['Frecuencia_Compras', 'Valor_Promedio_Orden']
    return X, y_true, feature_names
```

```python
def evaluate_clustering_performance(X, max_k=10):
    """
    Evalúa el rendimiento del clustering para diferentes valores de k.
    """
    ch_scores = []
    inertias = []
    k_range = range(2, max_k + 1)

    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)

        ch_score = calinski_harabasz_score(X, labels)
        ch_scores.append(ch_score)
        inertias.append(kmeans.inertia_)

        print(f"k={k:2d}: CH Index = {ch_score:7.2f}, Inercia = {kmeans.inertia_:7.2f}")
    
    return k_range, ch_scores, inertias
```

```python
def plot_clustering_results(X, y_true, k_range, ch_scores, inertias):
    """
    Genera visualizaciones del análisis de clustering.
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))

    # --- (1) Datos originales ---
    axes[0, 0].scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', alpha=0.7)
    axes[0, 0].set_title('Datos Originales (Clusters Verdaderos)')
    axes[0, 0].set_xlabel('Frecuencia de Compras')
    axes[0, 0].set_ylabel('Valor Promedio por Orden')

    # --- (2) Calinski–Harabasz Index vs k ---
    axes[0, 1].plot(k_range, ch_scores, 'bo-', linewidth=2, markersize=8)
    axes[0, 1].set_title('Calinski–Harabasz Index vs Número de Clusters')
    axes[0, 1].set_xlabel('Número de Clusters (k)')
    axes[0, 1].set_ylabel('CH Index')
    axes[0, 1].grid(True, alpha=0.3)

    optimal_k = k_range[np.argmax(ch_scores)]
    axes[0, 1].axvline(x=optimal_k, color='red', linestyle='--', alpha=0.7)
    axes[0, 1].text(optimal_k + 0.1, ch_scores[np.argmax(ch_scores)], 
                    f'k óptimo = {optimal_k}', fontsize=10, color='red')

    # --- (3) Método del Codo ---
    axes[1, 0].plot(k_range, inertias, 'ro-', linewidth=2, markersize=8)
    axes[1, 0].set_title('Método del Codo (Inercia vs k)')
    axes[1, 0].set_xlabel('Número de Clusters (k)')
    axes[1, 0].set_ylabel('Inercia')
    axes[1, 0].grid(True, alpha=0.3)

    # --- (4) Clustering con k óptimo ---
    kmeans_optimal = KMeans(n_clusters=optimal_k, random_state=42)
    labels_optimal = kmeans_optimal.fit_predict(X)

    axes[1, 1].scatter(X[:, 0], X[:, 1], c=labels_optimal, cmap='viridis', alpha=0.7)
    centroids = kmeans_optimal.cluster_centers_
    axes[1, 1].scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, linewidths=3)
    axes[1, 1].set_title(f'Clustering con k={optimal_k} (CH Index={max(ch_scores):.2f})')
    axes[1, 1].set_xlabel('Frecuencia de Compras')
    axes[1, 1].set_ylabel('Valor Promedio por Orden')

    plt.tight_layout()
    plt.show()

    return optimal_k
```

```python
def compare_algorithms(X):
    """
    Compara varios algoritmos de clustering usando Calinski–Harabasz Index.
    """
    algorithms = {
        'K-Means': KMeans(n_clusters=4, random_state=42),
        'Ward Linkage': AgglomerativeClustering(n_clusters=4, linkage='ward'),
        'Complete Linkage': AgglomerativeClustering(n_clusters=4, linkage='complete'),
        'Average Linkage': AgglomerativeClustering(n_clusters=4, linkage='average')
    }

    results = {}
    print("\nComparación de Algoritmos de Clustering:")
    print("-" * 50)

    for name, algorithm in algorithms.items():
        labels = algorithm.fit_predict(X)
        ch_score = calinski_harabasz_score(X, labels)
        results[name] = ch_score
        print(f"{name:15s}: CH Index = {ch_score:.2f}")

    best_algorithm = max(results, key=results.get)
    print(f"\nMejor algoritmo: {best_algorithm} (CH Index = {results[best_algorithm]:.2f})")

    return results
```

```python
# 1. Generar datos
print("1. Generando datos de clientes simulados...")
X, y_true, feature_names = create_customer_data(n_samples=800)

# 2. Normalizar
print("\n2. Normalizando datos...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Evaluar rendimiento con distintos k
print("\n3. Evaluando diferentes números de clusters:")
print("-" * 55)
k_range, ch_scores, inertias = evaluate_clustering_performance(X_scaled, max_k=10)

# 4. Visualizaciones
print("\n4. Generando visualizaciones...")
optimal_k = plot_clustering_results(X_scaled, y_true, k_range, ch_scores, inertias)

# 5. Comparación de algoritmos
print(f"\n5. Comparando algoritmos con k={optimal_k}:")
compare_algorithms(X_scaled)
```

```python
print(f"\n6. Análisis detallado para k={optimal_k}:")
print("-" * 40)

kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
labels_final = kmeans_final.fit_predict(X_scaled)
final_ch_score = calinski_harabasz_score(X_scaled, labels_final)

print(f"CH Index final: {final_ch_score:.2f}")
interpretacion = (
    "Excelente separación" if final_ch_score > 200 else
    "Buena separación" if final_ch_score > 100 else
    "Aceptable separación"
)
print(f"Interpretación: {interpretacion}")

# --- Estadísticas por cluster ---
print("\nEstadísticas por cluster (en escala original):")
for i in range(optimal_k):
    cluster_points = X[labels_final == i]
    print(f"Cluster {i}: {len(cluster_points)} puntos, "
          f"Centro = ({cluster_points.mean(axis=0)[0]:.2f}, {cluster_points.mean(axis=0)[1]:.2f})")
```
