# Normalized Mutual Information (NMI): Una Métrica Esencial para Clustering

```python
import os, warnings
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import normalized_mutual_info_score, confusion_matrix
from sklearn.preprocessing import StandardScaler

# --- Configuración general ---
os.environ["OMP_NUM_THREADS"] = "2"  # evita el warning de MKL en Windows
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak")

plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
np.random.seed(42)

print("=== ANÁLISIS DE CLUSTERING CON NORMALIZED MUTUAL INFORMATION (NMI) ===\n")
```

```python
# Generar datos con 4 clusters
n_samples = 300
n_centers = 4

X, y_true = make_blobs(
    n_samples=n_samples,
    centers=n_centers,
    cluster_std=1.2,
    center_box=(-10, 10),
    random_state=42
)

# Estandarizar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print(f"Datos generados: {X.shape[0]} muestras con {n_centers} clusters reales.")
```

```python
algorithms = {
    'K-Means': KMeans(n_clusters=4, random_state=42, n_init=10),
    'DBSCAN': DBSCAN(eps=0.5, min_samples=5),
    'Agglomerative': AgglomerativeClustering(n_clusters=4)
}

results = {}
predictions = {}

print("\nEvaluando modelos de clustering con NMI:\n")

for name, algorithm in algorithms.items():
    y_pred = algorithm.fit_predict(X_scaled)
    predictions[name] = y_pred
    
    nmi = normalized_mutual_info_score(y_true, y_pred)
    results[name] = nmi
    
    n_clusters_pred = len(np.unique(y_pred))
    n_noise = np.sum(y_pred == -1) if -1 in y_pred else 0
    
    print(f"{name}:")
    print(f"  • NMI Score: {nmi:.4f}")
    print(f"  • Clusters encontrados: {n_clusters_pred}")
    if n_noise > 0:
        print(f"  • Puntos de ruido: {n_noise}")
    print()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Comparación de Algoritmos de Clustering', fontsize=16, fontweight='bold')

# --- (1) Clusters reales ---
axes[0, 0].scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', alpha=0.7)
axes[0, 0].set_title('Ground Truth\n(Clusters Verdaderos)', fontweight='bold')
axes[0, 0].set_xlabel('Feature 1')
axes[0, 0].set_ylabel('Feature 2')
axes[0, 0].grid(True, alpha=0.3)

# --- (2-4) Predicciones de los algoritmos ---
positions = [(0, 1), (1, 0), (1, 1)]

for (row, col), (name, y_pred) in zip(positions, predictions.items()):
    # Manejar outliers (DBSCAN)
    colors = y_pred.copy()
    if -1 in colors:
        colors = np.where(colors == -1, max(colors) + 1, colors)
    
    axes[row, col].scatter(X[:, 0], X[:, 1], c=colors, cmap='viridis', alpha=0.7)
    axes[row, col].set_title(f'{name}\nNMI = {results[name]:.4f}', fontweight='bold')
    axes[row, col].set_xlabel('Feature 1')
    axes[row, col].set_ylabel('Feature 2')
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

algorithms_names = list(results.keys())
nmi_scores = list(results.values())

bars = plt.bar(
    algorithms_names, 
    nmi_scores, 
    color=['skyblue', 'lightcoral', 'lightgreen'], 
    edgecolor='black'
)

# Etiquetas en las barras
for bar, score in zip(bars, nmi_scores):
    plt.text(
        bar.get_x() + bar.get_width()/2, 
        bar.get_height() + 0.01, 
        f'{score:.3f}', 
        ha='center', va='bottom', fontweight='bold'
    )

plt.title('Comparación de NMI entre Algoritmos de Clustering', fontsize=14, fontweight='bold')
plt.ylabel('NMI Score', fontweight='bold')
plt.ylim(0, 1)
plt.grid(axis='y', alpha=0.3)
plt.axhline(y=0.8, color='red', linestyle='--', alpha=0.7, label='Clustering excelente (0.8)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
best_algorithm = max(results, key=results.get)
best_score = results[best_algorithm]

print(f"\n=== ANÁLISIS DEL MEJOR ALGORITMO: {best_algorithm} ===")
print(f"NMI Score: {best_score:.4f}\n")

if best_score >= 0.8:
    print("Clustering EXCELENTE — los clusters predichos coinciden casi perfectamente con los reales.")
elif best_score >= 0.6:
    print("Clustering BUENO — hay correspondencia clara entre clusters predichos y reales.")
elif best_score >= 0.4:
    print("Clustering MODERADO — existe cierta correspondencia, pero aún puede mejorarse.")
else:
    print("Clustering POBRE — poca correspondencia con la estructura real.")
```

```python
y_best = predictions[best_algorithm]
cm = confusion_matrix(y_true, y_best)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f'Matriz de Confusión — {best_algorithm}', fontweight='bold')
plt.ylabel('Clusters Verdaderos')
plt.xlabel('Clusters Predichos')
plt.tight_layout()
plt.show()
```
