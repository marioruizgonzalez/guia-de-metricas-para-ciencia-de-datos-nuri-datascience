# Precision Score: Evaluando la Exactitud de las Predicciones Positivas


```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report
import pandas as pd

# Configuración de gráficos
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("=== ANÁLISIS DE PRECISIÓN EN DETECCIÓN DE FRAUDE ===\n")
```

```python
# 1. CREAR DATASET SIMULANDO DETECCIÓN DE FRAUDE
print("=== CREANDO DATASET DE DETECCIÓN DE FRAUDE ===")

X, y = make_classification(
    n_samples=10000,
    n_features=20,
    n_informative=15,
    n_redundant=5,
    n_classes=2,
    weights=[0.95, 0.05],  # 5% fraudes
    random_state=42
)

# División train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Distribución del dataset:")
print(f"- Clase 0 (legítima): {(y_train == 0).sum()} muestras")
print(f"- Clase 1 (fraude): {(y_train == 1).sum()} muestras")
print(f"- Proporción de fraude: {(y_train == 1).mean():.3f}")
```

```python
# 2. ENTRENAR MÚLTIPLES MODELOS
print("\n=== ENTRENANDO MODELOS ===")

models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'SVM': SVC(random_state=42, probability=True)
}

results = {}
predictions = {}

for name, model in models.items():
    print(f"Entrenando {name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    predictions[name] = y_pred

    # Calcular métricas
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    results[name] = {
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

    print(f"  Precision: {precision:.3f}")
    print(f"  Recall: {recall:.3f}")
    print(f"  F1-Score: {f1:.3f}")
```

```python
# 3. CÁLCULO MANUAL DE PRECISIÓN
print("\n=== CÁLCULO MANUAL DE PRECISIÓN ===")

best_model = 'Random Forest'
y_pred_best = predictions[best_model]

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred_best)
tn, fp, fn, tp = cm.ravel()

print(f"Matriz de confusión para {best_model}:")
print(f"TN: {tn}, FP: {fp}, FN: {fn}, TP: {tp}")

# Cálculo manual
precision_manual = tp / (tp + fp)
recall_manual = tp / (tp + fn)

print(f"\nPrecision = TP / (TP + FP) = {tp} / ({tp} + {fp}) = {precision_manual:.3f}")
print(f"Recall = TP / (TP + FN) = {tp} / ({tp} + {fn}) = {recall_manual:.3f}")

# Verificación con sklearn
precision_sklearn = precision_score(y_test, y_pred_best)
print(f"Verificación con sklearn: {precision_sklearn:.3f}")
```

```python
# 4. VISUALIZACIONES PRINCIPALES
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Análisis de Precision Score en Detección de Fraude', fontsize=16, fontweight='bold')

# (1) Comparación de métricas entre modelos
ax1 = axes[0, 0]
df_results = pd.DataFrame(results).T
df_results.plot(kind='bar', ax=ax1, width=0.8)
ax1.set_title('Comparación de Métricas por Modelo')
ax1.set_ylabel('Score')
ax1.set_xlabel('Modelos')
ax1.legend(['Precision', 'Recall', 'F1-Score'])
ax1.set_ylim(0, 1)
plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

# (2) Matriz de confusión
ax2 = axes[0, 1]
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax2)
ax2.set_title(f'Matriz de Confusión - {best_model}')
ax2.set_ylabel('Valor Real')
ax2.set_xlabel('Predicción')
ax2.set_xticklabels(['Legítima', 'Fraude'])
ax2.set_yticklabels(['Legítima', 'Fraude'])

# (3) Distribución de predicciones
ax3 = axes[1, 0]
pred_counts = pd.Series(y_pred_best).value_counts()
colors = ['lightblue', 'salmon']
ax3.pie(pred_counts.values, labels=['Legítima', 'Fraude'], autopct='%1.1f%%', startangle=90, colors=colors)
ax3.set_title('Distribución de Predicciones del Modelo')

# (4) Análisis de errores
ax4 = axes[1, 1]
error_data = {'True Negatives': tn, 'False Positives': fp, 'False Negatives': fn, 'True Positives': tp}
bars = ax4.bar(error_data.keys(), error_data.values(), color=['lightgreen', 'red', 'orange', 'darkgreen'])
ax4.set_title('Análisis Detallado de Predicciones')
ax4.set_ylabel('Cantidad')
plt.setp(ax4.get_xticklabels(), rotation=45, ha='right')

# Añadir etiquetas
for bar in bars:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 5, f'{int(height)}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# 5. IMPACTO DE DIFERENTES THRESHOLDS EN PRECISIÓN
print("\n=== ANÁLISIS DE THRESHOLD Y SU IMPACTO EN PRECISIÓN ===")

rf_model = models['Random Forest']
y_proba = rf_model.predict_proba(X_test)[:, 1]

thresholds = np.arange(0.1, 1.0, 0.1)
precision_scores = []
recall_scores = []

for threshold in thresholds:
    y_pred_thresh = (y_proba >= threshold).astype(int)
    if (y_pred_thresh == 1).sum() > 0:
        precision = precision_score(y_test, y_pred_thresh)
        recall = recall_score(y_test, y_pred_thresh)
    else:
        precision, recall = 0, 0
    
    precision_scores.append(precision)
    recall_scores.append(recall)
    print(f"Threshold {threshold:.1f}: Precision={precision:.3f}, Recall={recall:.3f}")

# Gráfico Precision vs Recall
plt.figure(figsize=(10, 6))
plt.plot(thresholds, precision_scores, 'b-o', label='Precision', linewidth=2)
plt.plot(thresholds, recall_scores, 'r-s', label='Recall', linewidth=2)
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Trade-off entre Precision y Recall según Threshold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 6. REPORTE FINAL
print("\n=== REPORTE FINAL ===")
print(f"Mejor modelo: {best_model}")
print(f"Precision: {results[best_model]['precision']:.3f}")

print(f"Interpretación: De cada 100 transacciones predichas como fraude, "
      f"{results[best_model]['precision']*100:.1f}% realmente son fraudes.")

print(f"\nContexto bancario:")
print(f"- Falsos positivos: {fp}")
print(f"- Total de predicciones positivas: {(y_pred_best == 1).sum()}")
print(f"- Clientes legítimos bloqueados innecesariamente: {fp}")
print(f"- Tasa de error en predicciones positivas: {(fp/(tp+fp))*100:.1f}%")

print("\n=== CLASSIFICATION REPORT ===")
print(classification_report(y_test, y_pred_best, target_names=['Legítima', 'Fraude']))
```
