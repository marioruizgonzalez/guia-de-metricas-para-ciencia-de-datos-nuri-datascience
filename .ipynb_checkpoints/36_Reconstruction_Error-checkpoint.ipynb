# Reconstruction Error: Evaluando la Calidad en Reducción de Dimensionalidad

```python
import os, warnings
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Configuración general para Windows
os.environ["OMP_NUM_THREADS"] = "2"
warnings.filterwarnings("ignore", message="KMeans is known to have a memory leak")

plt.style.use('seaborn-v0_8')
np.random.seed(42)

print("=== ANÁLISIS PCA: RECONSTRUCTION ERROR Y VARIANZA EXPLICADA ===\n")
```

```python
# Cargar dataset de dígitos (8x8 píxeles = 64 características)
digits = load_digits()
X, y = digits.data, digits.target

print(f"Dataset cargado: {X.shape[0]} muestras, {X.shape[1]} características.")
print(f"Clases únicas: {np.unique(y)}")

# Dividir entre entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Estandarizar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Datos divididos: {X_train.shape[0]} entrenamiento / {X_test.shape[0]} prueba.")
```

```python
def calculate_reconstruction_error(X_original, X_reconstructed):
    """Calcula el Reconstruction Error promedio."""
    return np.mean(np.sum((X_original - X_reconstructed) ** 2, axis=1))

def calculate_relative_error(X_original, X_reconstructed):
    """Calcula el error relativo global."""
    return np.sum((X_original - X_reconstructed) ** 2) / np.sum(X_original ** 2)
```

```python
n_components_range = range(1, 65, 5)
train_errors = []
test_errors = []
relative_errors = []
explained_variance_ratios = []

print("\n=== EVALUANDO PCA PARA DIFERENTES NÚMEROS DE COMPONENTES ===")

for n_comp in n_components_range:
    pca = PCA(n_components=n_comp)
    X_train_reduced = pca.fit_transform(X_train_scaled)
    X_train_reconstructed = pca.inverse_transform(X_train_reduced)
    
    X_test_reduced = pca.transform(X_test_scaled)
    X_test_reconstructed = pca.inverse_transform(X_test_reduced)
    
    # Calcular errores
    train_error = calculate_reconstruction_error(X_train_scaled, X_train_reconstructed)
    test_error = calculate_reconstruction_error(X_test_scaled, X_test_reconstructed)
    rel_error = calculate_relative_error(X_test_scaled, X_test_reconstructed)
    
    train_errors.append(train_error)
    test_errors.append(test_error)
    relative_errors.append(rel_error)
    explained_variance_ratios.append(np.sum(pca.explained_variance_ratio_))
    
    print(f"{n_comp:2d} componentes -> Error Prueba: {test_error:.4f}, Varianza: {explained_variance_ratios[-1]:.3f}")
```

```python
# --- FIGURA 1: Gráficos de error y varianza ---
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("Análisis PCA: Reconstruction Error y Varianza Explicada", fontsize=16, fontweight='bold')

# --- Gráfico 1: Reconstruction Error ---
ax1.plot(n_components_range, train_errors, 'b-o', label='Entrenamiento', linewidth=2)
ax1.plot(n_components_range, test_errors, 'r-s', label='Prueba', linewidth=2)
ax1.set_title("Reconstruction Error vs Número de Componentes")
ax1.set_xlabel("Número de Componentes")
ax1.set_ylabel("Error de Reconstrucción")
ax1.legend()
ax1.grid(True, alpha=0.3)

# --- Gráfico 2: Error Relativo ---
ax2.plot(n_components_range, relative_errors, '^-', color='purple', linewidth=2)
ax2.set_title("Error Relativo vs Número de Componentes")
ax2.set_xlabel("Número de Componentes")
ax2.set_ylabel("Error Relativo")
ax2.grid(True, alpha=0.3)

# --- Gráfico 3: Varianza explicada ---
ax3.plot(n_components_range, explained_variance_ratios, color='orange', marker='o', linewidth=2)
ax3.axhline(y=0.95, color='red', linestyle='--', label='95% Varianza')
ax3.set_title("Varianza Explicada vs Número de Componentes")
ax3.set_xlabel("Número de Componentes")
ax3.set_ylabel("Varianza Explicada Acumulada")
ax3.legend()
ax3.grid(True, alpha=0.3)

# Dejar ax4 vacío o mostrar un texto resumen
ax4.axis("off")
ax4.text(0.1, 0.5,
         "Gráficos comparativos del rendimiento del PCA\n"
         "Basados en los errores de reconstrucción y varianza explicada.",
         fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()

# --- FIGURA 2: Comparación visual de reconstrucción ---
pca_example = PCA(n_components=10)
X_reduced_example = pca_example.fit_transform(X_train_scaled)
X_reconstructed_example = pca_example.inverse_transform(X_reduced_example)

sample_indices = [0, 1, 2, 3, 4]

fig, axes = plt.subplots(2, 5, figsize=(10, 4))
fig.suptitle("Reconstrucción Visual con 10 Componentes Principales", fontsize=14, fontweight='bold')

for i, idx in enumerate(sample_indices):
    axes[0, i].imshow(X_train[idx].reshape(8, 8), cmap='gray')
    axes[0, i].set_title(f'Original {y_train[idx]}')
    axes[0, i].axis('off')

for i, idx in enumerate(sample_indices):
    axes[1, i].imshow(X_reconstructed_example[idx].reshape(8, 8), cmap='gray')
    axes[1, i].set_title(f'Recons. {y_train[idx]}')
    axes[1, i].axis('off')

plt.tight_layout()
plt.show()
```

```python
print("=== ANÁLISIS DE RECONSTRUCTION ERROR ===\n")

optimal_components = n_components_range[np.argmin(test_errors)]
min_test_error = min(test_errors)
pca_optimal = PCA(n_components=optimal_components).fit(X_train_scaled)
variance_explained = np.sum(pca_optimal.explained_variance_ratio_)

print(f"Número óptimo de componentes: {optimal_components}")
print(f"Reconstruction Error mínimo: {min_test_error:.4f}")
print(f"Varianza explicada: {variance_explained:.3f} ({variance_explained*100:.1f}%)")
```

```python
print("\n=== ANÁLISIS DE TRADE-OFF ===")

for n_comp in [5, 10, 20, 30]:
    if n_comp in n_components_range:
        idx = list(n_components_range).index(n_comp)
        print(f"Componentes: {n_comp:2d} | "
              f"Error: {test_errors[idx]:.4f} | "
              f"Varianza: {explained_variance_ratios[idx]:.3f} | "
              f"Compresión: {(1 - n_comp/64)*100:.1f}%")
```

```python
results_df = pd.DataFrame({
    'Componentes': n_components_range,
    'Error_Entrenamiento': train_errors,
    'Error_Prueba': test_errors,
    'Error_Relativo': relative_errors,
    'Varianza_Explicada': explained_variance_ratios
})

print("\n=== TABLA RESUMEN (Primeras 5 filas) ===")
display(results_df.head().round(4))
```
